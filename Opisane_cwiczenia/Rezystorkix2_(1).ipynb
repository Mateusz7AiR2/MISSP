```
#Szeegowe połączenie rezystorów
from pulp import *
prob = LpProblem("Rezystoryszer",LpMaximize)
x1=LpVariable("i1")
x2=LpVariable("i2")
x3=LpVariable("i3")
x4=LpVariable("i4")
x5=LpVariable("i5")
#Szukamy maksymalny prad w układzie więc deklarujemy x1 - x5 (są to kolejne prądy)
#Używamy LpMaximize w celu znalezienia maksymlanej wartości
# Założenia zgodne z prawami Kirchoffa 
prob += x1 + x2, "maksymalny prąd"
prob += 8*x1 == 6*x2, "Napięcie R1 i R2"
prob += x1 + x2 == x3, "2"
prob += x3 == x4 + x5, "3"
prob += 10*x4 == 8*x5,  "napiecie na r4 i r5"
prob += x1 <= 2, "5"
prob += x2 <= 3, "6"
prob += x3 <= 4, "7"
prob += x4 <= 2, "8"
prob += x5 <= 2, "9"
#Rozwiazanie problemu i wypisanie rozwiazania
prob.writeLP("Rezystoryszer")
prob.solve()
print(prob.solve())
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("max prad= ", value(prob.objective))
```

```
#Mostek 
from pulp import *
prob = LpProblem("Most",LpMinimize)
# Szukanie minimalnych strat na rezystorach
#Deklaracja pradów, które bedą w zakresie -100 - 1000
x1=LpVariable("i1", -100, 100,LpContinuous)
x2=LpVariable("i2", -100, 100,LpContinuous)
x3=LpVariable("i3", -100, 100,LpContinuous)
x4=LpVariable("i4", -100, 100,LpContinuous)
x5=LpVariable("i5", -100, 100,LpContinuous)
#Założenia wynikające z tabeli ograniczeń 

prob += x1 <=5, "deltax1"
prob += x1 >=3, "-deltax1"

prob += x2 <=3, "deltax2"
prob += x2 >=1, "-deltax2"

prob += x3 <=3, "deltax3"
prob += x3 >=1, "-deltax3"

prob += x4 <=3, "deltax4"
prob += x4 >=1, "-deltax4"

prob += x5 <=5, "deltax5"
prob += x5 >=3, "-deltax5"
#Założenia wynikające z praw Kirchoffa
prob += x1 - x3 - x4 == 0, "prady 1"
prob += x2 + x3 - x5 == 0, " prady 2"
prob += x1 + x2 == x4 + x5 , "prady 3"
prob += 6 * x1 + 10 * x2 + 4 * x3 + 7 * x4 +3 * x5, "moce na oporniku"
#Rozwiązanie problemu i wypisanie wyników
prob.writeLP("Most.lp")
prob.solve()
print(prob.solve())
print("Status:", LpStatus[prob.status])

for v in prob.variables():
    print(v.name, "=", v.varValue)
print("straty= ", value(prob.objective)/1000)

```
