# Zadanie:

Opracuj program dzięki, któremu:
1.   Za minimalną stawkę obsadzisz wymaganą ilość pracowników w ciągu danego dnia.
2.   Każdy pracownik może pracować w max 3 dni w tygodniu




| Dni tygodnia| ilość pracowników|
|------------|-------------------|
| Poniedzialek    | 2     |
| Wtorek     | 1       |
| Środa    | 1         |
| Czwartek      | 1  |
| Piątek | 3                |

| Pracownicy| Stawka za godzine|Kiedy mogą pracować|
|------------|-------------------|-----------------|
| Ania   | 100   | Każdy dzień|
| Stefan     | 50  |Poniedzialek|
| Hektor   | 60   |Pon, wt, śr|
| Olaf      | 40 |Każdy dzień|
| Lidia | 110  |Czw, piątek|
| Piotr| 70  |Pon, wt, śr|

Zadanie wykonane:


```
from pulp import *
# Dla każdego pracownika tworzę listę dni w których może pracować( pon - pt) 0 oznacza iż tego dnia nie praacuje natomiast 1 iż przychodzi do pracy 
prob = LpProblem("Team_pracownikow",LpMinimize)
# Używam LpMinimize by wyszukiwane zostało rozwiązanie które spełni 1 podpunkt założeń (obsada za najniższą stawkę)
ponAnia=LpVariable("ponAnia", 0, 1)
wtAnia=LpVariable("wtAnia", 0, 1)
srAnia=LpVariable("srAnia", 0, 1)
czwAnia=LpVariable("czwAnia", 0, 1)
ptAnia=LpVariable("ptAnia", 0, 1)

ponStefan=LpVariable("ponStefan",0 , 1)
wtStefan=LpVariable("wtStefan",0 , 1)
srStefan=LpVariable("srStefan",0 , 1)
czwStefan=LpVariable("czwStefan",0 , 1)
ptStefan=LpVariable("ptStefan",0 , 1)

ponHektor=LpVariable("ponHektor",0 , 1)
wtHektor=LpVariable("wtHektor",0 , 1)
srHektor=LpVariable("srHektor",0 , 1)
czwHektor=LpVariable("czwHektor",0 , 1)
ptHektor=LpVariable("ptHektor",0 , 1)

ponOlaf=LpVariable("ponOlaf",0 , 1)
wtOlaf=LpVariable("wtOlaf",0 , 1)
srOlaf=LpVariable("srOlaf",0 , 1)
czwOlaf=LpVariable("czwOlaf",0 , 1)
ptOlaf=LpVariable("ptOlaf",0 , 1)

ponLidia=LpVariable("ponLidia",0 , 1)
wtLidia=LpVariable("wtLidia",0 , 1)
srLidia=LpVariable("srLidia",0 , 1)
czwLidia=LpVariable("czwLidia",0 , 1)
ptLidia=LpVariable("ptLidia",0 , 1)

ponPiotr=LpVariable("ponPiotr",0 , 1)
wtPiotr=LpVariable("wtPiotr",0 , 1)
srPiotr=LpVariable("srPiotr",0 , 1)
czwPiotr=LpVariable("czwPiotr",0 , 1)
ptPiotr=LpVariable("ptPiotr",0 , 1)
#Obliczam koszta obsady sklepu w ciągu tygodnia
prob += (ponAnia+ wtAnia+ srAnia + czwAnia + ptAnia)*100 + (ponStefan + wtStefan + srStefan + czwStefan + ptStefan) * 50 + (ponHektor + wtHektor + srHektor + czwHektor + ptHektor)*60 +(ponOlaf + wtOlaf + srOlaf + czwOlaf + ptOlaf)* 40 +  (ponLidia + wtLidia + srLidia + czwLidia + ptLidia)* 110 + (ponPiotr + wtPiotr + srPiotr + czwPiotr + ptPiotr)*70,  "pencja pracownikow"
#Wprowadzam ograniczenia max 3 dni pracy dla pracownika oraz jego możliwości podjęcia pracy w danym dniu
prob += (ponAnia+ wtAnia+ srAnia + czwAnia + ptAnia) <=3, " max 3 dni robi Ania"

prob += (ponStefan + wtStefan + srStefan + czwStefan + ptStefan) <=3, "max 3 dni robi Stefcio"
prob += (wtStefan + srStefan + czwStefan + ptStefan) <=0, "czas kiedy nie robi Stefcio"

prob += (ponHektor + wtHektor + srHektor + czwHektor + ptHektor) <=3, " max 3 dni robi Hektor"
prob += (czwHektor + ptHektor) <=0, " czas kiedy nie robi Hektor"

prob += (ponOlaf + wtOlaf + srOlaf + czwOlaf + ptOlaf) <=3, " max 3 dni robi Olaf"

prob += (ponLidia + wtLidia + srLidia + czwLidia + ptLidia) <=3, " max 3 dni robi Lidia"
prob += (ponLidia + wtLidia + srLidia) <=0, " czas kiedy nie robi Lidia"

prob += (ponPiotr + wtPiotr + srPiotr + czwPiotr + ptPiotr) <=3, " max 3 dni robi Piotr"
prob += (czwPiotr + ptPiotr) <=0, " czas kiedy nie robi Piotr"
#sprawdzam który pracownik przyjdzie do pracy w danym dniu
prob += (ponAnia + ponStefan + ponHektor + ponOlaf + ponLidia + ponPiotr) ==2, "Pracownicy w pon"
prob += (wtAnia + wtStefan + wtHektor + wtOlaf + wtLidia + wtPiotr) ==1, "Pracownicy w wt"
prob += (srAnia + srStefan + srHektor + srOlaf + srLidia + srPiotr) ==1, "Pracownicy w sr"
prob += (czwAnia + czwStefan + czwHektor + czwOlaf + czwLidia + czwPiotr) ==1, "Pracownicy w czw"
prob += (ptAnia + ptStefan + ptHektor + ptOlaf + ptLidia + ptPiotr) ==3, "Pracownicy w pt"
#Zaciągam rozwiązanie i wypisuje je poniżej
prob.solve()
print(prob.solve())
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Total kasa= ", value(prob.objective))
```
