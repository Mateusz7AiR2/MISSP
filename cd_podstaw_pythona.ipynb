### Własne typy

```python
zajecia = [["wf"],
           ["informatyka", "nauka czytania"],
           ["relax", "jedzenie"],
           ["mispy","mispy w domu"],
           ["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"]
          ]

dnityg= ["Pon","Wto","Sro","Czwar","Pio"]

def dodaj_dwie_liczby(liczba1,liczba2):
    return liczba1+liczba2
```

```python
class DzienZajec():
    """obiekt trzymający dany dzień zajęć i przypisane mu metody"""
    
    
    def __init__(self, lista_na_dany_dzien, nazwa_dnia): #definicja
        self.lista_zajec= lista_na_dany_dzien
        self.dzien = nazwa_dnia
        

    
    def ile_mispow(self):
        ile_m = sum([1 for zajecie in self.lista_zajec if "misp" in zajecie.lower()])
        return ile_m

    @property #zamienia na wlasciwosc   
    def ile_zajec(self):
        return len(self.lista_zajec)
    
    def __repr__(self):
        zajecia_do_stringa = ", ".join(self.lista_zajec)
        return "W dzien {} mam zajecia: {}".format(self.dzien, zajecia_do_stringa)
    
```

```python
pio_zajecia = DzienZajec(["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"],
                        "piatek")
```

Stworz typ NumerTelefonu

 -- przy tworzeniu sprawdza czy właściwy numer (Polski) jeśli nie pisze, że error (ew. Exception)
 
 -- gdy printuję to mi pisze polskimi słowami i.e. "jeden, pięć, siedem, zero"
 
 -- gdy wywołuję właściwość wojewodzwtwo, podaje mi wojewodztwo ew. ze to komorkowy
 
 

```python
# ma na takie rzeczy być w stanie odpowiedzieć
class SprawdzNumer():
  def __init__(self,sprawdzany_numer):
    self.numer_telefonu = sprawdzany_numer
    if len(self.numer_telefonu) == 9:
      return print("Prawidłowy_numer_domowy")
    else:
      return print("To_nie_jest_domowy_numer")
  @property
  def miasta(self):
     zmienna_kierunkowa = self.numer_telefonu[0]+ self.numer_telefonu[1]
     if zmienna_kierunkowa == "42":
        return print("Miasto Łódz") 
     else:   
        if zmienna_kierunkowa  == "24":
          return print("Miasto Płock")
        else:
          if zmienna_kierunkowa == "12" :
            return print("Miasto Kraków")
          else:
            return print("komurkowy")
  @property
  def pisemnie_numery(self):
    cyfry_słownie = [ "zero", "jeden", "dwa", "trzy", "cztery", "piec", "szesc", "siedem", "osiem", "dziewiec" ]
    zapisany_numer = " "
    for i in range(len(self.numer_telefonu)):
      zapisany_numer = zapisany_numer +" "+cyfry_słownie[int(self.numer_telefonu[i])]
    return zapisany_numer
sprawdzany_numer = SprawdzNumer("42345679")
sprawdzany_numer.miasta
sprawdzany_numer.pisemnie_numery
```
